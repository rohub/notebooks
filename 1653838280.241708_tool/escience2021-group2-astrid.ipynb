# Abstract 

The Arctic Ocean ecosystems are experiencing rapid changes. This could have possible implications for aerosol and cloud feedbacks that are not currently represented in CMIP6 models. One important proxy for plankton life is oceanic chlorophyll. This study investigates the representation of chlorophyll in two CMIP6 models; GFDL-ESM4 and MPI-ESM1-2-LR, and compares it with satellite observations. The chlorophyll concentration elvovement in time is investigated in Bering Strait and Fram Strait. The satellite observations show a positive trend of 0.1 mg m$^{-3}$year$^{-1}$ (Fram Strait) and 0.06 mg m$^{-3}$year$^{-1}$ (Bering Strait) in maximum chlorophyll values. The projected changes in SSP2-4.5 in GFDL-ESM4 is a negative trend of -0.018 mg m$^{-3}$year$^{-1}$ in maximum values, while MPI-ESM1-2-LR shows a positive trend of 0.011 mg m$^{-3}$year$^{-1}$ in maximum values. Chlorophyll is compared with the other biological proxies bacterial carbon concentration and dissolved organic carbon. Chlorophyll is found to have a moderate correlation of 0.53 with bacterical carbon in GFDL-ESM4. No correlation is found with dissolved organic carbon, 0.09 for GFDL-ESM4 and 0.05 for MPI-ESM1-2-LR. GFDL-ESM4 and MPI-ESM1-2-LR also show diverging relationship between sea ice concentration and chlorophyll, where chlorophyll decreases along with declining sea ice at the end of SSP2-4.5, while chlorophyll increases in MPI-ESM1-2-LR with declining sea ice.

# Introduction

The Arctic is warming at one of the highest rates in the world. The Arctic sea ice decline is accelerating the warming, and is also affecting many ecosystems dramatically. One of the ecosystem components that are being dramatically affected is the algal life we find at the floor of the Arctic Ocean food chain. The role this life plays in the ocean ecology and carbon cycles is key, but it is also thought to play a significant role in other Earth system feedbacks, some of which we are still far from gaining a comprehensive understanding of.

These lifeforms are called plankton, and consist of both the plant phytoplankton, the micro-animal zooplankton and the bacterial plankton, with a larger diversity exisiting within the three groups. Phytoplankton create energy through photosynthesis, and make up food for zoo-plankton and other small animals, which again are an important food source for other parts of the ecosystem. When plankton sediment, the carbon in the plankton is stored in the deep ocean or ocean floor, and such they act as carbon sink. Phytoplankton also produce dimethylsulfide (DMS), a sulfuric gas that can contribute to cloud condensation particles. However, plankton as well make up a large part of the biological matter located at the surface of the ocean. When suspended into the air through sea spray, this matter is a source of primary biological aerosols in the Arctic that can act as ice nucleating particles. In other words, the plankton life forms can have multiple effects on aerosol concentrations, clouds and cloud properties in addition to their role in the ecosystem. In pristine areas such as the Arctic, these could make up an important part of the aerosol populations. 

The sea ice decline is likely to play a role both for the plankton populations themselves as well as how plankton populations affect aerosols. Lack of sea ice can make the plankton more available to the atmosphere and thereby increase the aerosol flux. In addition, light availability is important for the phytoplankton, and less sea ice will lead to more available light, and the decline of sea ice can therefore lead to increase in phytoplankton populations. Indeed, this is what has been observed so far in the Arctic {cite}`scott2020`. However, sea ice is also thought to provide a nutrient-rich layer for plankton growth when it melts. Studies have found chlorophyll in summer to be positively correlated with thick ice coverage in the previous winter {cite}`kyung2014`. There are also observations of large plankton blooms under sea ice {cite}`gradinger1996`. This suggest that sea ice can play a dual role for plankton both as a nutrient provider and a sun blocker - a life facilitator and a life prohibitor. What it mainly suggests is perhaps that the optimal life conditions for plankton is one where these two roles of the sea ice are balanced. Less sea ice could increase productivity, but too little sea ice could hinder it. How the sea ice affects the plankton population will therefore play a complex role in how Arctic aerosol populations are changed in a warming climate.

However, we know little about to what extent the plankton populations are affecting aerosol populations today, and therefore little about what effects changes in the plankton population will have. In order to understand the role that plankton will play for the climate and the marine ecosystems in a warming Arctic, earth system models (ESMs) are a valuable tool. However, even though ocean biogeochemistry models are becoming increasingly complex, there is little representation of dynamic connections between ocean biogeochemistry and the rest of the climate {cite}`seferian2020`. 

This study aims to investigate how CMIP6 earth system models are representing plankton populations in a changing climate, and how this representation compares to observations. The models investigated here are GFDL-EMS4 and MPI-ESM1-2-LR. We use chlorophyll as proxy for plankton, as phytoplankton contain chlorophyll, and phytoplankton are an integral part of the food chain for other species. However, chlorophyll is not a perfect proxy for plankton and biological matter as other species are not representented as well. The study therefore also contains a comparison between chlorophyll and the other biological proxies bacterial carbon concentration and dissolved organic carbon. Finally, we study how chlorophyll concentrations and sea ice change together in a warming climate within both models. The study thereby aims to show how different models portray the complex relationship between sea ice and plankton life, and what groundwork needs to be done in the modelling world before building further complexity in the form of aerosol-ocean life interactions.

# Results

In figure 2 we see little change in summer chlorophyll concentration between the periods 2002-2010 and 2010-2018 in the satellite data. the white area in the middle, where the chlorophyll is not measurable due to the presence of sea ice seems to have become slightly smaller, and the Fram Strait seems to have a darker streak of green. There are high concentrations of chlorophyll measured by the coastal areas, which are not seen in the models. While GFDL-ESM4 has very high values of chlorophyll in the Bering Strait and higher values in the Central Arctic than in the Atlantic waters, MPI-ESM1-2-LR has generally high values of chlorophyll, and lowest values in the central Arctic. 

In figure 3 we see that there is a general increase in the maximum values of chlorophyll concentrations in Bering Strait and Fram Strait, and an increase in the Fram Strait yearly mean values as well in the satellite observations. GFDL-ESM4 greatly overestimates the chlorophyll in the Bering Strait, and captures the Fram Strait concentration better inspite some year-to-year under- and overestimations. MPI-ESM1-2-LR genereally overestimates the chlorophyll concentration in both straits, though less so than GFDL-ESM4 in the Bering Strait. MPI-ESM1-2-LR appears to have larger year-to-year fluctuations than GFDL-ESM4. 

In figure 4 we see that the trends in GFDL-ESM4 and MPI-ESM1-2-LR diverge. GFDL-ESM4 shows negative trends in both mean amd maximum values in both Bering Strait and Fram Strait. The strongest trend is seen in the maximum values, where there is a trend of -0.018 mg m$^{-3}$year$^{-1}$ in Bering Strait and -0.005 mg m$^{-3}$year$^{-1}$ in Fram Strait. On the other hand, MPI-ESM1 shows mainly positive trends. In the maximum values, MPI-ESM1-2-LR shows a trend of 0.011 mg m$^{-3}$year$^{-1}$ in Bering Strait and 0.003 mg m$^{-3}$year$^{-1}$ in Fram Strait. In Bering Strait the trend of the mean is also positive, but the mean trend in Fram Strait is interestingly slightly negative with -0.001 mg m$^{-3}$year$^{-1}$.

In figure 5 we see that the chlorophyll and bacterical carbon, and to some degree the dissolved organic carbon, have the same pattern around Svalbard in GFDL-ESM4. # Discussion


The models we have studied appear to treat chlorophyll very differently. While GFDL-ESM4 shows a decrease in concentrations in SSP2-4.5, MPI-ESM1-2-LR mostly shows the opposite, and GFDL-ESM4 also shows less chlorophyll with less sea ice. An immediate question is whether this decline is somehow related to the sea ice itself, or if it caused by something else. It has been suggested that the stratification caused by warming of the upper ocean will lead to less nutrient cycling from the deep, and thereby less producitivity in the upper ocean {cite}`behrenfeld2006`. The decline in chlorophyll that we see in GFDL-ESM4 corresponding to the decline in sea ice infigure 7 could therefore also be caused indirectly by the sea ice decline, as the sea decline leads to more radiation being absorbed by the upper ocean. In MPI-ESM1-2-LR the increase in light availability seems to conquer any lack of nutrient availability that may or may not be present in the model, leading to an increase in chlorophyll. However, the chlorophyll concentrations do not continue to increase above the 2.5 mg m$^{-3}$ already present in the 2015-2025 period in figure 7, indicating that the chlorophyll concentrations may be nutrient limited to some extent. There seems to be quite a different method for handling chlorophyll concentration below sea ice in GFDL-ESM4 and MPI-ESM1-2-LR as the chlorophyll concentrations below sea ice covered area is higher in GFDL-ESM4 than in MPI-ESM1-2-LR in figure 2. Which of the models are most realistic is an open question, as we don't know what is hiding beneath the sea ice covered areas that are not registred by the satellite observations in figure 2. 

The specific response that plankton have to decline in sea ice can be very varying across plankton species, and we cannot expect that CMIP6 models capture these correctly. Most models have 2-3 plankton types, but this is represents the diversity in plankton very crudely {cite}`henson2021`. 

The different responses we could see in aerosol populations and further climate feedbacks are as diverging as the plankton responses. More plankton and less sea ice cover could lead to more areosols, which could lead to more clouds and more reflected radiation. However, more available ice nucleation particles could lead to more ice content in clouds, which could lead to less reflective and short-lived clouds, and thus less reflected radiation. If the plankton concentrations would decrease with less sea ice (and warmer oceans) this could lead to less available aerosols from ocean life, and opposite. This field consists of unknown feedbacks, and requires further study.

It is also important to remark that we have mainly considered the chlorophyll response to sea ice decline and warming oceans in this study. As we have seen, the chlorophyll does not correlate well with other biological proxies. While bacterial carbon is a better indicator of ocean life than dissolved organic carbon, both can matter for the aerosol flux. The relations we have studied between chlorophyll and sea ice decline might not be giving us a full picture for the future ocean life and aerosol interactions.
