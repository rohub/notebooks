```python
filename= "C:/Work/RELIANCE_cnr_ISMAR/Monitoring_Alberoni_WWF2022/monitoring_items_mb.csv"

file = 'monitoring_items_mb.csv'
filename = os.path.join('./', file)
```

```python
#Store the data from the file in a variable
data = infile.read()
```

```python
import pandas as pd
df = pd.read_csv(filename, sep = ';')
# Print the Data
df.head()
```

### 3- Descriptive statistics  

```python
#Summary statistics by waste category
df.groupby("WasteCategory")['Quantity'].describe().reset_index()
```

### 4- Plot data 

#### 4.1-  Plastic waste category  

```python
# Plot
kwargs = dict(hist_kws={'alpha':.6}, kde_kws={'linewidth':2})
x1 = df.loc[df.WasteCategory =='PLASTICS', 'Quantity']

plt.figure(figsize=(8,5), dpi= 80)
sns.distplot(x1, color="g", label='PLASTICS', **kwargs)
plt.gca().set(title='Histogram density of Plastic waste quantity (10/04/2022)/ Alberoni ')

plt.legend();
plt.savefig('monitoring_PLASTCS_waste.png')
```
