```python
a = adam.Auth()

a.setKey(adam_key)
a.setAdamCore('https://reliance.adamplatform.eu')
a.authorize()
```

Exploration of ADAM catalogue

```python
from adamapi import Datasets

datasets = Datasets(a)
catalogue = datasets.getDatasets()

#Extracting the size of the catalogue

total = catalogue['properties']['totalResults']
items = catalogue['properties']['itemsPerPage']
pages = total//items

print('----------------------------------------------------------------------')
print('\033[1m' + 'List of available datasets:')
print ('\033[0m')

#Extracting the list of datasets across the whole catalogue

for i in range(0,pages):
    page = datasets.getDatasets(page = i)
    for element in page['content']:
        print(element['title'] + "\033[1m" + " --> datasetId "+ "\033[0m" + "= " + element['datasetId'])
```

```python
datasetID1 ='90697:MODh20PARYR_4km'
datasetID2 ='69615:MODh20kd490YR_4km'

print('\033[1;34m' + 'Metadata of ' + datasetID1 + ':')
print ('\033[0;0m')

paged = datasets.getDatasets(datasetID1)
for i in paged.items():
    print("\033[1m" +  str(i[0]) + "\033[0m" + ': ' + str(i[1]))
    
long_name = paged['title']

print('\033[1;34m' + 'Metadata of ' + datasetID2 + ':')
print ('\033[0;0m')

paged = datasets.getDatasets(datasetID2)
for i in paged.items():
    print("\033[1m" +  str(i[0]) + "\033[0m" + ': ' + str(i[1]))
    
long_name = paged['title']
```

Dataset subsetting

```python
from adamapi import Search
from geojson_rewind import rewind
import json

# The geometry field is extracted from a GeoJSON object , retrieving the value of the "feature" element

geom = """{ "type": "Feature", "properties": { }, "geometry": { "type": "Polygon", "coordinates": [ [ [ 12.0, 44.5 ], [ 12.0, 45.8], [ 13.82, 45.8], [ 13.82, 44.5 ],[ 12.0, 44.5 ] ] ] } }
"""

# The GeoJson object needs to be rearranged according to the counterclockwise winding order.This operation is executed ih the next few lines to obtain a geometry that
# meets the requirements of the method. Geom_1 is the final result to be used in the discovery operation.

output = rewind(geom)

geom_dict = json.loads(output)

geom_1 = str(geom_dict['geometry'])



search = Search( a )
results = search.getProducts(
    datasetID1, 
    geometry= geom_1,
    startDate = '2002-01-01',
    endDate = '2018-01-01'
 )

#Printing the results

print('\033[1m' + 'List of available products:')
print ('\033[0m')
count = 1
for i in results['content']:

        print("\033[1;31;1m" + "#" + str(count))
        print ('\033[0m')
        for k in i.items():
            print(str(k[0]) + ': ' + str(k[1]))
        count = count+1
        print('------------------------------------')
        
        search = Search( a )
results = search.getProducts(
    datasetID2, 
    geometry= geom_1,
    startDate = '2002-01-01',
    endDate = '2018-12-01'
 )

#Printing the results

print('\033[1m' + 'List of available products:')
print ('\033[0m')
count = 1
for i in results['content']:

        print("\033[1;31;1m" + "#" + str(count))
        print ('\033[0m')
        for k in i.items():
            print(str(k[0]) + ': ' + str(k[1]))
        count = count+1
        print('------------------------------------')
```
